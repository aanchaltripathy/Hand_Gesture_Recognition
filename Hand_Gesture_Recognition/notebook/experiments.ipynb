```python
# üñêÔ∏è Hand Gesture Recognition - Experiments
# Data exploration and model testing
#import ==========
import pickle
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
import mediapipe as mp
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ===============================
# loading gesture names
# ===============================
with open('../gesture.names', 'r') as f:
    classNames = f.read().strip().split('\n')

print("Gestures:", classNames)

# ===============================
# loading dataset
# ===============================
with open('../data.pickle', 'rb') as f:
    data = pickle.load(f)

print("Type of data:", type(data))
print("Number of samples:", len(data))

# Extract X and y properly
y = [item[0] for item in data]
X = [item[1] for item in data]

X = np.array(X, dtype=np.float32)
y = np.array(y)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Unique labels:", set(y))

# ===============================
# visualizing some hand landmarks
# ===============================
mpHands = mp.solutions.hands
mpDraw = mp.solutions.drawing_utils

plt.figure(figsize=(5,5))
for i in range(3):
    landmarks = np.array(X[i])
    plt.scatter(landmarks[:,0], landmarks[:,1], label=classNames[y[i]])
    plt.title(f"Gesture: {classNames[y[i]]}")
    plt.gca().invert_yaxis()
    plt.legend()
    plt.show()

# ===============================
# loading TensorFlow SavedModel
# ===============================
model = tf.saved_model.load('../mp_hand_gesture')
infer = model.signatures['serving_default']
print("‚úÖ Model loaded successfully")
print("Output keys:", list(infer.structured_outputs.keys()))

# ===============================
# test prediction on a sample
# ===============================
sample = X[0].reshape(1, -1)
output_key = list(infer.structured_outputs.keys())[0]
prediction = infer(tf.convert_to_tensor(sample))[output_key].numpy()
pred_class = classNames[np.argmax(prediction)]
print("True class:", classNames[y[0]])
print("Predicted class:", pred_class)

# ===============================
# evaluate full dataset accuracy
# ===============================
preds = []
for i in range(len(X)):
    sample = tf.convert_to_tensor([X[i].flatten()])
    output = infer(sample)[output_key].numpy()
    preds.append(np.argmax(output))

acc = accuracy_score(y, preds)
print(f"\n‚úÖ Overall Model Accuracy: {acc*100:.2f}%")

print("\nüìã Classification Report:")
print(classification_report(y, preds, target_names=classNames))

# Confusion matrix
cm = confusion_matrix(y, preds)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=classNames, yticklabels=classNames)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Gesture Recognition')
plt.show()

# ===============================
# gesture Distribution
# ===============================
df = pd.DataFrame({'gesture': [classNames[i] for i in y]})
sns.countplot(data=df, x='gesture')
plt.title('Gesture Distribution')
plt.xticks(rotation=30)
plt.show()

print("\n‚úÖ Experiment complete!")
```
